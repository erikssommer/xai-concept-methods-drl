```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Plot the sigmoid fuction

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 100)
y = sigmoid(x)

plt.plot(x, y)
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid()

# Print the equation of the sigmoid function on the plot inside a text box with a white background and black border in the upper left corner
plt.text(-10, 0.9, r'$\sigma(x) = \frac{1}{1 + e^{-x}}$', fontsize=12, bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5'))

plt.show()
```

```python
# Plot the ReLU function
def relu(x):
    return np.maximum(0, x)

x = np.linspace(-10, 10, 100)
y = relu(x)

plt.plot(x, y)
plt.title('ReLU Function')
plt.xlabel('x')
plt.ylabel('relu(x)')
plt.grid()

# Print the equation of the ReLU function on the plot inside a text box with a white background and black border in the upper left corner
plt.text(-10, 9, r'${ReLU}(x) = \max(0, x)$', fontsize=12, bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5'))

plt.show()
```
