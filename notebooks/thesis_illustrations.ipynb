```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Plot the sigmoid fuction

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 1000)
y = sigmoid(x)

plt.plot(x, y)
#plt.title('Sigmoid Function')
plt.xlabel('x', fontsize=14)
plt.ylabel(r'$\sigma(x)$', fontsize=14)
plt.grid()

# Increase the font size of the x and y axis labels
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Make them one decimal place
plt.gca().xaxis.set_major_formatter('{:.1f}'.format)
plt.gca().yaxis.set_major_formatter('{:.1f}'.format)

# Print the equation of the sigmoid function on the plot inside a text box with a white background and black border in the upper left corner
plt.text(-10, 0.9, r'$\sigma(x) = \frac{1}{1 + e^{-x}}$', fontsize=14, bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5'))

plt.show()
```

```python
# Plot the ReLU function
def relu(x):
    return np.maximum(0, x)

x = np.linspace(-1, 1, 1000)
y = relu(x)

plt.plot(x, y)
#plt.title('ReLU Function')
plt.xlabel('x', fontsize=14)
plt.ylabel('ReLU(x)', fontsize=14)
plt.grid()

# Increase the font size of the x and y axis labels
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Make them one decimal place
plt.gca().xaxis.set_major_formatter('{:.1f}'.format)
plt.gca().yaxis.set_major_formatter('{:.1f}'.format)

# Print the equation of the ReLU function on the plot inside a text box with a white background and black border in the upper left corner
plt.text(-1, 0.9, r'${ReLU}(x) = \max(0, x)$', fontsize=14, bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5'))

plt.show()
```
